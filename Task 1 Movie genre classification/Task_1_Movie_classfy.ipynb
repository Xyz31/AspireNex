```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import plotly.express as px

tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(train_data['Description'])
y_train = train_data['Genre']
X_test_tfidf = tfidf_vectorizer.transform(test_data['Description'])
y_test = test_data_solution['Genre']
```

### 3D visualize

```python
# Perform PCA
pca = PCA(n_components=3)
X_pca_3d = pca.fit_transform(X_train_tfidf.toarray())

# Create a DataFrame for Plotly
df_3d = pd.DataFrame(
    data=X_pca_3d,
    columns=['PC1', 'PC2', 'PC3']
)
df_3d['Class'] = y_train

# Create an interactive 3D scatter plot
fig = px.scatter_3d(
    df_3d,
    x='PC1',
    y='PC2',
    z='PC3',
    color='Class',
    title='3D Scatter Plot of PCA Features Colored by Class'
)

```

```python
import pandas as pd
# Perform PCA to reduce dimensionality to 2D
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_train_tfidf.toarray())

# Create a DataFrame for Plotly
df_2d = pd.DataFrame(
    data=X_pca_2d,
    columns=['PC1', 'PC2']
)
df_2d['Class'] = y_train

# Create an interactive 2D scatter plot
fig = px.scatter(
    df_2d,
    x='PC1',
    y='PC2',
    color='Class',
    title='2D Scatter Plot of PCA Features Colored by Class'
)

# Show the plot
fig.show()
```

```python
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, mean_absolute_error, mean_squared_error, roc_auc_score, roc_curve, auc, confusion_matrix



from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
import warnings
warnings.filterwarnings('ignore')

from sklearn.naive_bayes import MultinomialNB
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_tfidf, y_train)
y_test_pred_nb = nb_classifier.predict(X_test_tfidf)

print("Naive Bayes - Accuracy on test set:", accuracy_score(y_test, y_test_pred_nb))
print("Naive Bayes - Classification Report on test set:\n", classification_report(y_test, y_test_pred_nb))
```

#### Model 2

```python
import warnings
warnings.filterwarnings('ignore')


logreg_classifier = LogisticRegression(max_iter=1000)
logreg_classifier.fit(X_train_tfidf, y_train)
y_test_pred_logreg = logreg_classifier.predict(X_test_tfidf)

print("Logistic Regression - Accuracy on test set:", accuracy_score(y_test, y_test_pred_logreg))
print("Logistic Regression - Classification Report on test set:\n", classification_report(y_test, y_test_pred_logreg))
```

#### Model 3

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV

svm_classifier = LinearSVC(C=0.1, max_iter=1000, dual=False)

# Fit the model
svm_classifier.fit(X_train_tfidf, y_train)

# Predict
y_test_pred_svm = svm_classifier.predict(X_test_tfidf)

print("SVM - Accuracy on test set:", accuracy_score(y_test, y_test_pred_svm))
print("SVM - Classification Report on test set:\n", classification_report(y_test, y_test_pred_svm))
```

#### Model 4

```python
import warnings
warnings.filterwarnings('ignore')


from sklearn.ensemble import RandomForestClassifier

# Initialize and train Random Forest model
rf_classifier = RandomForestClassifier(n_estimators=70, max_depth=100,random_state=42)
rf_classifier.fit(X_train_tfidf, y_train)

# Predict on test data
y_test_pred_rf = rf_classifier.predict(X_test_tfidf)

# Evaluate the model
print("Random Forest - Accuracy on test set:", accuracy_score(y_test, y_test_pred_rf))
print("Random Forest - Classification Report on test set:\n", classification_report(y_test, y_test_pred_rf))
```

#### Model 5

```python
import warnings
warnings.filterwarnings('ignore')


import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report

# Encode the genre labels
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)

# Convert labels to one-hot encoding
y_train_one_hot = to_categorical(y_train_encoded)
y_test_one_hot = to_categorical(y_test_encoded)

# Build the neural network model
nn_model = Sequential()
nn_model.add(Dense(128, input_dim=5000, activation='relu'))  # number of neurons
nn_model.add(Dropout(0.5))
nn_model.add(Dense(64, activation='relu'))  # number of neurons
nn_model.add(Dropout(0.5))
nn_model.add(Dense(len(np.unique(y_train_encoded)), activation='softmax'))

# Compile the model
nn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
nn_model.fit(X_train_tfidf.toarray(), y_train_one_hot, epochs=20, batch_size=64, validation_split=0.2, verbose=2)

# Predict on test data
y_test_pred_nn = nn_model.predict(X_test_tfidf.toarray())
y_test_pred_nn_labels = encoder.inverse_transform(np.argmax(y_test_pred_nn, axis=1))

# Evaluate
print("Neural Network - Accuracy on test set:", accuracy_score(y_test, y_test_pred_nn_labels))
print("Neural Network - Classification Report on test set:\n", classification_report(y_test, y_test_pred_nn_labels))
```
